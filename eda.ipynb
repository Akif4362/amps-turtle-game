```python
mean_age = []
for i in range(3):
    x = df[df['pclass'] == i+1]['age'].mean()
    mean_age.append(x)
    print(f'mean age of class {i+1} is {x}')
```

```python
df[df['pclass'] == 1]['age'].fillna(38.23)
df[df['pclass'] == 2]['age'].fillna(29.87)
df[df['pclass'] == 3]['age'].fillna(25.14)
```

```python
df.info()
```

```python
mapping = {i +1 : mean_age[i] for i in range(3)}
mapping
```

```python
df['age'] = df['age'].fillna(df['pclass'].map(mapping))
```

```python
missing = []
for i in df.columns:
    x = df[i].isnull().sum()
    missing.append(x)
pd.DataFrame(missing, index=df.columns, columns=['missing values'])
```
