## Density and Contour Plots

```python
def f(x, y):
    return np.sin(x) ** 10 + np.sin(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors="purple");
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.cm.RdGy
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar();
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy', interpolation='gaussian', aspect='equal')
plt.colorbar();
```

## How to Choose the Right Plot?

| Scenario | Best Plot Type |
|----------|--------------|
| Trend over time | **Line plot** |
| Relationship between two variables | **Scatter plot** |
| Comparing categories | **Bar plot** |
| Data distribution | **Histogram** |
| Detecting outliers | **Box plot** |
| Part-to-whole comparison | **Pie chart** (use sparingly) |
| Heat or correlation visualization | **Heatmap** |
| 3D function representation | **Contour / 3D plot** |


## Choosing the Right Plot Based on Data Types

| Data Type Combination | Best Plot Type |
|----------------------|---------------|
| **Categorical vs. Categorical** | **Stacked Bar Chart**, **Grouped Bar Chart**, **Heatmap** |
| **Categorical vs. Numerical** | **Box Plot**, **Violin Plot**, **Bar Chart** |
| **Numerical vs. Numerical** | **Scatter Plot**, **Line Plot** (for trends), **Heatmap** (for correlations) |
| **Numerical (distribution analysis)** | **Histogram**, **Box Plot**, **Violin Plot** |
| **Multiple Categorical + Numerical** | **Facet Grid**, **Grouped Bar Chart**, **Stacked Histogram** |
| **Time Series (Time vs. Numerical)** | **Line Plot**, **Area Plot**, **Scatter Plot** |
| **3 Numerical Variables** | **3D Scatter Plot**, **Bubble Chart**, **Contour Plot** |
| **Geospatial Data** | **Choropleth Map**, **Scatter Plot on Maps**, **Heatmap** |

## Multiple Subplots

```python
x = np.linspace(0, 10, 1000)

plt.subplot(2, 3, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 3, 2)
plt.plot(x, np.cos(x))
plt.subplot(2, 3, 3)
plt.plot(x, x)
plt.subplot(2, 3, 4)
plt.plot(x, x**2)
plt.subplot(2, 3, 5)
plt.plot(x, x**3)
plt.subplot(2, 3, 6)
plt.plot(x, x**(1/2.0));
```

### Annotation

```python
plt.style.use("default")

fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
           arrowprops=dict(facecolor='red', shrink=0.05))

ax.annotate('local minima', xy=(5 * np.pi, -1), xytext=(2, -6),
           arrowprops=dict(arrowstyle='->', connectionstyle="angle3, angleA=0, angleB=-90"));
```

## 3D Plotting

```python
from mpl_toolkits import mplot3d
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")

fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
ax = plt.axes(projection='3d')

zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline)

zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap="Greens");
```

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 40, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
